```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 5000

data = {
    "student_id": range(1, n + 1),
    "cgpa": np.round(np.random.uniform(5.0, 10.0, n), 2),
    "attendance": np.random.randint(50, 101, n),
    "coding_score": np.random.randint(0, 101, n),
    "aptitude_score": np.random.randint(0, 101, n),
    "communication_score": np.random.randint(0, 101, n),
    "projects_count": np.random.randint(0, 8, n),
    "internships_count": np.random.randint(0, 4, n),
    "hackathon_count": np.random.randint(0, 6, n)
}

df = pd.DataFrame(data)

placement_score = (
    df["cgpa"] * 10 +
    df["coding_score"] * 0.4 +
    df["aptitude_score"] * 0.3 +
    df["communication_score"] * 0.1 +
    df["projects_count"] * 3 +
    df["internships_count"] * 5
)

df["placement_status"] = (placement_score > 120).astype(int)

df.to_csv("../datasets/student_data.csv", index=False)

print("Dataset created successfully!")
print(df.head())
```
